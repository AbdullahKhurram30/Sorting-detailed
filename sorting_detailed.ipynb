# Iteration vs. recursion

A [Fibonacci Sequence](https://en.wikipedia.org/wiki/Fibonacci_number) is a list of numbers, in which each number is the sum of the two previous ones (starting with 0 and 1). Mathematically, we can write it as:
- F(0) = 0
- F(1) = 1
- F(n) = F(n-1) + F(n-2) for n larger than 1.

Your task is to compute the **n**th number in the sequence, where **n** is an input to the function.

Use the code skeletons below to **provide two solutions** to the problem: an **iterative**, and a **recursive** one. For the recursive solution, feel free to use a helper function or add keyword arguments to the given function.

````python
def fibonacci_iterative(n):
    """
    Returns the corresponding value in the fibonacci sequence for a number n
    via iterative approach
     
    Parameters
    ----------
    int: n for which we need the corresponding value in the fibonacci sequence

    Returns
    -------
    int: the number in the fibonacci sequence
    """
    #check if n = 0
    if n == 0:
        return 0
    #if n >= 1, move forward with the iteration
    else:
        #declaring the variables
        x = 0
        fib = 1
        #for loop to iterate from 1 to n
        for i in range(1, n):
            #change the temporary variable to the sum of two previous numbers
            temp = fib + x
            #for i > 2, increment x by 1 starting from 1
            x = fib
            #change the fib variable to the temporary sum
            fib = temp
        #return the final value for the fibonacci number    
        return fib

def fibonacci_recursive(n):
    """
    Returns the corresponding value in the fibonacci sequence for a number n
    via recursive approach
     
    Parameters
    ----------
    int: n for which we need the corresponding value in the fibonacci sequence

    Returns
    -------
    int: the number in the fibonacci sequence
    """
    #check if n = 1 or 0 since they have the same values as n(base case)
    if n <= 1:
        return n
    #call the function recursively until the function returns the needed value
    else:
        #gives the sum of the fibonacci for n-1 and n-2
        return fibonacci_recursive(n - 1) + fibonacci_recursive(n - 2)
    
assert fibonacci_iterative(0) == 0
assert fibonacci_recursive(0) == 0
assert fibonacci_iterative(4) == 3
assert fibonacci_recursive(4) == 3
````

The algorithm works by selecting the first element in the array. It then compares the selected element with all the
elements in the array. Until it finds an element that is greater than or equal to the selected value, it keeps incrementing 
intended position variable which is the position our selected element is supposed to be in. Once it finds a number

a number that is equal to the selected element, it swaps the number with the position before it to keep the algorithm
stable.
Next, it nowselects the element that we just swapped by comparing it with the elements until it finds it's 
correct position. Once there are no more swaps it increments the for loop and performs the same step on the 
subarray[1:]. This is done until there are no more swaps and the check is complete. Once the algorithm has checked
all the elements in the list to make sure that they are sorted, it returns the sorted array.

**(a)** Use the following pseudocode to implement `merge_3_sort()`. It is similar to merge sortonly that instead of splitting the initial array into halves, you will now split it into thirds, call the function recursively on each sublist, and then merge the triplets together. You might want to refer to this [beautiful resource](https://drive.google.com/file/d/1XH5bNiHhVchVWoCDLBtvjsCFkL9cU0do/view?pli=1) written by Prof. Drummond for details about the regular merge sort algorithm.

````python
def merge_3_sort(array, p, q):
    """
    Sorts array[p] to array[q] in place.
    E.g., to sort an array A, we will run 
    `merge_3_sort(A, 0, len(A)-1)`.


    Parameters
    ----------
    array : Python list or numpy array
    p : int
        index of array element to start sorting from
    q : int
        index of last array element to be sorted

    Returns
    -------
    array: a sorted Python list

    """
    #check if the array has values
    if p < q:
        
        #Find the index for dividing the array
        n1 = int((q - p)/3 + p)
        n2 = int(q - ((q - p)/3))
    
        #Call the function recursively on the subarrays divided based on indices calculated above
        merge_3_sort(array, p, n1)
        merge_3_sort(array, n1 + 1, n2)
        merge_3_sort(array, n2 + 1, q)
        
        #Combine the sorted arrays to get the final solution
        merge_3(array, p, n1, n2, q)
    
    return array

    
def merge_3(array, p, q, r, s):
    """
    Merges 3 sorted sublists 
    (array[p] to array[q], array[q+1] to array[r] and array[r+1] to array[s]) 
    in place.
    
    Parameters
    ----------
    array : Python list or numpy array
    p : int
        index of first element of first sublist
    q : int
        index of last element of first sublist
    r : int
        index of last element of second sublist
    s : int
        index of last element of third sublist

    """
    #initialize the subarrays and then add infinity to the end as we will need them to act as "sentinels"
    left = array[p:q + 1] + [float('inf')]
    middle = array[q + 1:r + 1] + [float('inf')]
    right = array[r + 1:s + 1] + [float('inf')]
    
    #initializing the indices for the three subarrays to compare them with values from other subarrays
    i = 0
    j = 0
    k = 0

    #running a for loop to compare all the values in various indices and get a sorted array
    for t in range(p, s + 1):
        #seeing if the compared values are smallest for the left subarray and then changing the t-th index to the value
        if min(left[i], middle[j], right[k]) == left[i]:
            array[t] = left[i]
            i += 1

        #seeing if the compared values are smallest for the moiddle subarray and then changing the t-th index to the value
        elif min(left[i], middle[j], right[k]) == middle[j]:
            array[t] = middle[j]
            j += 1
            
        #seeing if the compared values are smallest for the right subarray and then changing the t-th index to the value
        elif min(left[i], middle[j], right[k]) == right[k]:
            array[t] = right[k]
            k += 1
    
    return array
````

Analytical solution

As shown above, the order of growth of Merge_3_sort and Selection sort is O(n*log(3)(n)) and O(n^2) respectively. This means that even if selection sort performs better than merge sort initially, the order of growth will lead it to always perform worse after a certain number of inputs. However, merge sort will perform worse initially because of the number of operations it performs on small inputs. We do not have defined computational constants so we can not calculate the exact inflection point analytically but we can find it experimentally.

Graphical solution

To find the inflection point, we should run the worst case scenarios for increasing input sizes on both selection and merge sort functions. Since merge sort works on the same complexity for all cases, it doesn't have a defined worst case so we can just run them both on the worst case for selection sort. These would be reversed arrays. We can plot a graph of these measurements and find the point where selection sort becomes more costly than the merge sort.

````python
import timeit
import matplotlib.pyplot as plt

def selection_sort(array):
    """"
    Sorts the given array by finding the minumum value of the 
        unsorted subarray and moving it to the right position.
    
    Parameters
    -------
    array: List of unsorted numbers
    
    Returns
    -------
    array: Sorted python list
    
    """
    
    n = len(array)
    
    for i in range(0, n):
        minimum = i
        for j in range(i + 1, n):
            if array[j] < array[minimum]:
                minimum = j
        array[i], array[minimum] = array[minimum], array[i]
        
    return array
        
def selection_sort_time(array):
    """
    Returns
    -------
    Time for running the function
    """
    #start time
    time1 = timeit.default_timer()
    
    #calling the selection sort function
    selection_sort(array)
    
    #end time
    time2 = timeit.default_timer()
    
    return time2 - time1

for a in range(10):

    #initializing the test case list
    test = []
    test_b = []

    #making the test cases
    for a in range(1, 75):
        test.append(list(range(a, 0, -1)))
        test_b.append(list(range(a, 0, -1)))

    #initializing the lists for time taken to sort the test cases    
    test_time = []
    selection_test_time = []

    #runs the merge sort function on the test cases and appends time taken to the list
    for t in test:
        time1 = timeit.default_timer()
        merge_3_sort(t, 0, len(t) - 1)
        time2 = timeit.default_timer()
        test_time.append(time2 - time1)

    #runs the selection sort function on the test cases and appends the time taken to the list    
    for t in test_b:
        selection_test_time.append(selection_sort_time(t))
````

````python
x_axis_values = [i for i in range(1, 75)]

#making the line graph for the selection sort vs merge sort function
plt.plot(x_axis_values, test_time)
plt.plot(x_axis_values, selection_test_time, alpha = 0.5)
plt.xlabel('n')
plt.ylabel('Time taken')
plt.show()
````

Looking at the graphs above, we can observe the time taken for both the algorithms to sort the cases inserted into them.
````python
def merge_3_sort_threshold(array, p, q):
    #Condition to call selection sort
    if q - p <= 45:
        
        #selection sort
        for i in range(0, len(array) - 1):
            minimum = i

            for j in range(i + 1, len(array)):
                 if  array[j] < array[minimum]:
                           minimum = j

            if minimum != i:
                # Swap the values
                 array[minimum], array[i] = array[i], array[minimum]
        
    else:
        #Define the split points within array
        n1 = int((q - p)/3 + p)
        n2 = int(q - ((q - p)/3))
    
        #Recursively sort the three subarrays
        merge_3_sort(array, p, n1)
        merge_3_sort(array, n1 + 1, n2)
        merge_3_sort(array, n2 + 1, q)
        
        #Call merge_3 function to combine the partially sorted subarrays
        merge_3(array, p, n1, n2, q)
    
    return array
    
def merge_3(array, p, q, r, s):
    left = array[p:q + 1] + [float('inf')]
    middle = array[q + 1:r + 1] + [float('inf')]
    right = array[r + 1:s + 1] + [float('inf')]
    
    #initializing the indices for the three subarrays to compare them with values from other subarrays
    i = 0
    j = 0
    k = 0

    #running a for loop to compare all the values in various indices and get a sorted array
    for t in range(p, s + 1):
        #seeing if the compared values are smallest for the left subarray and then changing the t-th index to the value
        if min(left[i], middle[j], right[k]) == left[i]:
            array[t] = left[i]
            i += 1

        #seeing if the compared values are smallest for the moiddle subarray and then changing the t-th index to the value
        elif min(left[i], middle[j], right[k]) == middle[j]:
            array[t] = middle[j]
            j += 1
            
            #seeing if the compared values are smallest for the right subarray and then changing the t-th index to the value
        elif min(left[i], middle[j], right[k]) == right[k]:
            array[t] = right[k]
            k += 1
    
    return array
````
